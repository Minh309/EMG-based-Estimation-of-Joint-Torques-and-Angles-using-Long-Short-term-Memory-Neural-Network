```python
%matplotlib inline
import numpy as np
import pandas as pd
import scipy as sp
import matplotlib.pyplot as plt

setting = pd.read_csv('setting.csv', header = 0, usecols = [0,1])
EMG_features = setting.iloc[0,1]
IK = setting.iloc[1,1]
ID = setting.iloc[2,1]
output_name = setting.iloc[4,1]
output_path = setting.iloc[3,1]
```

```python
df_EMG = pd.read_csv(EMG_features, header=0)
df_EMG.describe()
```

```python
df_IK = pd.read_csv(IK, header=0, usecols = [10,11,17,18])
df_IK.describe()
```

```python
df_ID = pd.read_csv(ID, header=0, usecols = [16,17,18,19])
df_ID.describe()
```

```python
data_frame = pd.concat([df_EMG,df_IK,df_ID],axis = 1)
data_frame.describe()
```

```python
output_file = output_path + output_name + '.csv'
data_frame.to_csv(output_file, index = False)
```
